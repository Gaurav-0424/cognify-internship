Q1. Analyze the distribution of aggregate ratings and determine the most common rating range.

```python

# Step 1: Analyze the distribution of aggregate ratings
plt.figure(figsize=(8, 6))

# Plot the histogram for aggregate ratings
sns.histplot(df['Aggregate rating'], kde=True, color='blue', bins=10)
plt.title('Distribution of Aggregate Ratings', fontsize=14, fontweight='bold')
plt.xlabel('Aggregate Rating', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xlim(0, 5)  # Ratings range from 0 to 5
plt.show()

# Step 2: Determine the most common rating range
rating_bins = pd.cut(df['Aggregate rating'], bins=[0, 1, 2, 3, 4, 5], right=False)
most_common_range = rating_bins.value_counts().idxmax()

# Print the most common rating range
print(f"The most common aggregate rating range is: {most_common_range}")

# Step 3: Calculate the average number of votes received by restaurants
average_votes = df['Votes'].mean()
```

Q1. Identify the most common combinations ofcuisines in the dataset

```python
from collections import Counter

# Assuming the cuisine column is named 'Cuisines'
cuisine_series = df['Cuisines'].dropna()  # Drop rows with missing cuisine data

# Split cuisine strings into lists of individual cuisines
cuisine_lists = cuisine_series.str.split(', ')

# Create a list of cuisine combinations (tuples) with at least two cuisines
cuisine_combinations = []
for cuisines in cuisine_lists:
    if isinstance(cuisines, list) and len(cuisines) >= 2:
        cuisine_combinations.append(tuple(sorted(cuisines)))  # Sort to ensure consistent ordering

# Count the frequency of each cuisine combination
combination_counts = Counter(cuisine_combinations)

# Get the most common combinations
most_common_combinations = combination_counts.most_common(10)  # Adjust the number as needed

# Print the results
print("Most Common Cuisine Combinations (at least 2 cuisines):")
for combination, count in most_common_combinations:
    print(f"{', '.join(combination)}: {count} times")

# --- Create Horizontal Bar Chart ---
combinations = [', '.join(comb) for comb, count in most_common_combinations]
counts = [count for comb, count in most_common_combinations]

# Set up the color palette
palette = sns.color_palette("coolwarm", len(combinations))

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(combinations, counts, color=palette)

# Add data labels to the bars
for bar in bars:
    width = bar.get_width()  # Get the width of the bar (the frequency count)
    plt.text(width + 1, bar.get_y() + bar.get_height() / 2,  # Position the label just to the right of the bar
             str(width), va='center', fontsize=10, color='black')  # Add the frequency count as the label

# Add labels and title
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Cuisine Combinations', fontsize=12)
plt.title('Top 10 Cuisine Combinations', fontsize=14, fontweight='bold')

# Invert y-axis to show the most frequent on top
plt.gca().invert_yaxis()

# Display the plot
plt.tight_layout()
plt.show()
```

Task 3: Geographic Analysis

Q1. Plot the locations of restaurants on a map using longitude and latitude coordinates.

```python
import folium
from folium.plugins import MarkerCluster

# Assuming 'df' contains the longitude and latitude data for restaurants
# Replace 'Longitude' and 'Latitude' with the actual column names in your DataFrame
# Example: df['Longitude'] and df['Latitude']

# Step 1: Create the base map centered around the average coordinates
avg_lat = df['Latitude'].mean()  # Calculate the average latitude
avg_lon = df['Longitude'].mean()  # Calculate the average longitude

# Create a base map centered around the average coordinates
restaurant_map = folium.Map(location=[avg_lat, avg_lon], zoom_start=12, control_scale=True)

# Step 2: Initialize the MarkerCluster to group nearby restaurants
marker_cluster = MarkerCluster().add_to(restaurant_map)

# Step 3: Add each restaurant location to the map
for index, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"Restaurant: {row['Restaurant Name']}Rating: {row['Aggregate rating']}",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(marker_cluster)

# Step 4: Save the map to an HTML file (for interactive viewing)
restaurant_map.save('restaurant_map.html')

# Step 5: Display the map inline (optional for Jupyter Notebooks)
restaurant_map
```
